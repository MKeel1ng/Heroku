```

import ipywidgets as widgets
from numpy import average
import os 
from PIL import Image
from random import choice
from random import random
#%cd "/content/drive/My Drive/stylegan2-ada/stylegan2-ada"
#pathHKB='./out/cell/gen0/'
#pathFIB='./out/cell/gen1/'
#pathRealHKB='./raw/cells512/mk512/'
#pathRealFIB='./raw/cells512/mf512/'

pathHKB='./images/fake/HKB/'
pathFIB='./images/fake/FIB/'
pathRealHKB='./images/real/HKB/'
pathRealFIB='./images/real/FIB/'
```

###Cell type classifier

I've trained a StyleGan2 model on two classes consisting of microscopy images of human cells. The classes are fibroblasts and epidermal keratinocytes. Try guessing for each cell which class it belongs to.

*Note: Images where captured at different magnifications by different users with different staining protocols. Keratinocyte images were trained with 1/5th of the number of images as Fibroblast images and contain several significant drug treatments that modify their shape and make-up. Training is not complete*



```
btn_run = widgets.Button(description='Show')
btn_fib = widgets.Button(description='Fibroblast')
btn_hkb = widgets.Button(description='Keratinocyte')
btn_reset_class = widgets.Button(description='Reset Score')
btn_result_class=widgets.Button(description='Show My Score')
out_plclass = widgets.Output()


resbox_class=widgets.IntText(
    value=0,
    description='% Score',
    disabled=False
)

truth=widgets.Text(
    value='None', 
    disabled=False
)

correct=[]

def on_click_Generate_class(change):

  out_plclass.clear_output()

  #set seed

  if random()<0.5:
    filesFake=[]
    for filename in os.listdir(pathFIB):
      filesFake.append(filename)
    im = Image.open(pathFIB+choice(filesFake))
    truth.value='fib'
  else:
    filesFake=[]
    for filename in os.listdir(pathHKB):
      filesFake.append(filename)
    im = Image.open(pathHKB+choice(filesFake))
    truth.value='hkb'

    


  with out_plclass:
      display(im)


def on_click_fib(change):
  if truth.value=='fib':
    correct.append(1)
  else:
    correct.append(0)
  on_click_Generate_class(True)

def on_click_hkb(change):
  if truth.value=='hkb':
    correct.append(1)
  else:
    correct.append(0)
  on_click_Generate_class(True)

def on_click_result(change):
    try:
      correct_score=average(correct)*100
      resbox_class.value=correct_score
    except:
      resbox_class.value=0

def on_click_reset(change):
    resbox_class.value=0
    global correct    
    correct=[]

#btn_run.on_click(on_click_Generate)
on_click_Generate_class(True)
btn_fib.on_click(on_click_fib)
btn_hkb.on_click(on_click_hkb)
btn_result_class.on_click(on_click_result)
btn_reset_class.on_click(on_click_reset)

widgets.VBox([widgets.HBox([btn_fib,btn_hkb]),widgets.HBox([btn_result_class,btn_reset_class]),resbox_class,out_plclass])
```

##Real or fake?

Fibroblasts are better trained at the moment but please do try both cell types!

```
cell_type=widgets.Dropdown(
    options=['Fibroblast', 'Keratinocyte'],
    value='Fibroblast',
    description='What cell do you want to generate?:',
    disabled=False,
)
btn_reset = widgets.Button(description='Reset Score')
btn_run = widgets.Button(description='Show')
btn_left = widgets.Button(description='I\'m real')
btn_right = widgets.Button(description='No I\'m real!')
btn_result=widgets.Button(description='Show My Score')
out_pl1 = widgets.Output()
out_pl2 = widgets.Output()

resbox=widgets.IntText(
    value=0,
    description='% Score',
    disabled=False
)

truth=widgets.Text(
    value='None', 
    disabled=False
)

correct=[]

def on_click_Generate(change):

  out_pl1.clear_output()
  out_pl2.clear_output()
  #set seed

  if cell_type.value=='Fibroblast':
    filesFake=[]
    for filename in os.listdir(pathFIB):
      filesFake.append(filename)
    imFake = Image.open(pathFIB+choice(filesFake))

    filesReal=[]
    for filename in os.listdir(pathRealFIB):
      filesReal.append(filename)
    imReal = Image.open(pathRealFIB+choice(filesReal))

  else:
    filesFake=[]
    for filename in os.listdir(pathHKB):
      filesFake.append(filename)
    imFake = Image.open(pathHKB+choice(filesFake))

    filesReal=[]
    for filename in os.listdir(pathRealHKB):
      filesReal.append(filename)
    imReal = Image.open(pathRealHKB+choice(filesReal))

  if random()>0.5:
    truth.value='left'
    with out_pl1:
      display(imReal)
    with out_pl2:
      display(imFake)

  else:
    truth.value='right'
    with out_pl1:
      display(imFake)
    with out_pl2:
      display(imReal)


def on_click_LEFT(change):
  if truth.value=='left':
    correct.append(1)
  else:
    correct.append(0)
  on_click_Generate(True)

def on_click_RIGHT(change):
  if truth.value=='right':
    correct.append(1)
  else:
    correct.append(0)
  on_click_Generate(True)

def on_click_result(change):
    try:
      correct_score=average(correct)*100
      resbox.value=correct_score
    except:
      resbox.value=0

def on_change(change):
    on_click_Generate(True)

def on_click_reset(change):
    resbox.value=0
    global correct    
    correct=[]


cell_type.observe(on_change)
#btn_run.on_click(on_click_Generate)
on_click_Generate(True)
btn_left.on_click(on_click_LEFT)
btn_right.on_click(on_click_RIGHT)
btn_result.on_click(on_click_result)
btn_reset.on_click(on_click_reset)



widgets.VBox([cell_type,widgets.HBox([btn_result,resbox, btn_reset]), widgets.HBox([widgets.VBox([btn_left,out_pl1]),widgets.VBox([btn_right,out_pl2])])])
```
