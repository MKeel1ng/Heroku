```

import ipywidgets as widgets
from numpy import average
import os 
from PIL import Image
from random import choice
from random import random
#%cd "/content/drive/My Drive/stylegan2-ada/stylegan2-ada"
#pathHKB='./out/cell/gen0/'
#pathFIB='./out/cell/gen1/'
#pathRealHKB='./raw/cells512/mk512/'
#pathRealFIB='./raw/cells512/mf512/'

pathHKB='./images/fake/HKB/'
pathFIB='./images/fake/FIB/'
pathRealHKB='./images/real/HKB/'
pathRealFIB='./images/real/FIB/'
```

##Cell classifier

Fibroblasts are better trained at the moment but please do try both cell types!

```
cell_type=widgets.Dropdown(
    options=['Fibroblast', 'Keratinocyte'],
    value='Fibroblast',
    description='What cell do you want to generate?:',
    disabled=False,
)

btn_run = widgets.Button(description='Show')
btn_left = widgets.Button(description='I\'m real')
btn_right = widgets.Button(description='No I\'m real!')
btn_result=widgets.Button(description='Show My Score')
out_pl1 = widgets.Output()
out_pl2 = widgets.Output()

resbox=widgets.IntText(
    value=0,
    description='Score out of 100%',
    disabled=False
)

truth=widgets.Text(
    value='None', 
    disabled=False
)

correct=[]

def on_click_Generate(change):

  out_pl1.clear_output()
  out_pl2.clear_output()
  #set seed

  if cell_type.value=='Fibroblast':
    filesFake=[]
    for filename in os.listdir(pathFIB):
      filesFake.append(filename)
    imFake = Image.open(pathFIB+choice(filesFake))

    filesReal=[]
    for filename in os.listdir(pathRealFIB):
      filesReal.append(filename)
    imReal = Image.open(pathRealFIB+choice(filesReal))

  else:
    filesFake=[]
    for filename in os.listdir(pathHKB):
      filesFake.append(filename)
    imFake = Image.open(pathHKB+choice(filesFake))

    filesReal=[]
    for filename in os.listdir(pathRealHKB):
      filesReal.append(filename)
    imReal = Image.open(pathRealHKB+choice(filesReal))

  if random()>0.5:
    truth.value='left'
    with out_pl1:
      display(imReal)
    with out_pl2:
      display(imFake)

  else:
    truth.value='right'
    with out_pl1:
      display(imFake)
    with out_pl2:
      display(imReal)


def on_click_LEFT(change):
  if truth.value=='left':
    correct.append(1)
  else:
    correct.append(0)
  on_click_Generate(True)

def on_click_RIGHT(change):
  if truth.value=='right':
    correct.append(1)
  else:
    correct.append(0)
  on_click_Generate(True)

def on_click_result(change):
    correct_score=average(correct)*100
    resbox.value=correct_score

#btn_run.on_click(on_click_Generate)
on_click_Generate(True)
btn_left.on_click(on_click_LEFT)
btn_right.on_click(on_click_RIGHT)
btn_result.on_click(on_click_result)

widgets.VBox([widgets.HBox([cell_type,btn_result,resbox]), widgets.HBox([widgets.VBox([btn_left,out_pl1]),widgets.VBox([btn_right,out_pl2])])])
```
